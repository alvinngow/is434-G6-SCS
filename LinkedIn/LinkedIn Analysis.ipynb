```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
# import .csv
df = pd.read_csv("scs linkedin posts.csv")
```

```python
df["postUrl"].groupby(df["postActualDate"]).count()
df.head()
```

```python
df["postActualDate"].unique().size
```

```python
df["postActualDate"] = pd.to_datetime(df["postActualDate"], format='%d/%m/%Y')
df = df.sort_values(by="postActualDate")
```

```python
# plt.plot(xAxis,yAxis)
plt.plot(df["postActualDate"].unique(), df["postUrl"].groupby(df["postActualDate"]).count())
plt.rcParams["figure.figsize"] = (15,5)
plt.title("SCS's No. of LinkedIn Posts by Time")
plt.xlabel('Time')
plt.ylabel('Number of Posts')

# plt.bar(df["likeCount"].groupby(df["postActualDate"]).mean(), width=2000, height=1000)
# plt.tick_params(axis='y')
plt.show()
```

```python
import matplotlib.ticker as ticker

x1 = df["postActualDate"].unique()
d1 = df["likeCount"].groupby(df["postActualDate"]).mean()
t1 = df["postUrl"].groupby(df["postActualDate"]).count()

fig, ax1 = plt.subplots(figsize=(10, 6))
color = '#00bbe4'
ax1.set_title("SCS's No. of LinkedIn Posts and Likes by Time", fontsize=16)
ax1.set_xlabel('Time', fontsize=16)
ax1.set_ylabel("Average No. of Likes", fontsize=16, color=color)
ax1.bar(x1, height=d1, width=0.9, color=color)
ax1.tick_params(axis='y', colors=color)

ax2 = ax1.twinx()  # share the x-axis, new y-axis
color = 'crimson'
ax2.set_ylabel("No. of LinkedIn Posts", fontsize=16, color="red")
ax2.plot(x1, t1, color=color)
ax2.tick_params(axis='y', colors=color)

plt.show()
```
