```python
import pandas as pd
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```python
# import .csv
df = pd.read_csv("scs linkedin posts.csv")
df.head()
```

```python
df["neg"] = int()
df["neu"] = int()
df["pos"] = int()
df["compound"] = int()
```

```python
df
```

```python
# append sentiment analysis data to df
for idx, row in df.iterrows():
    sia.polarity_scores("Wow, NLTK is really powerful!")
#     df.loc[idx].at["neg"] = sia.polarity_scores(row["textContent"])["neg"]
#     df.loc[idx].at["pos"] = sia.polarity_scores(row["textContent"])["pos"]
#     df.loc[idx].at["neu"] = sia.polarity_scores(row["textContent"])["neu"]
#     df.loc[idx].at["compound"] = sia.polarity_scores(row["textContent"])["compound"]
    df.at[idx, "neg"] = sia.polarity_scores(row["textContent"])["neg"]
    df.at[idx, "pos"] = sia.polarity_scores(row["textContent"])["pos"]
    df.at[idx, "neu"] = sia.polarity_scores(row["textContent"])["neu"]
    df.at[idx, "compound"] = sia.polarity_scores(row["textContent"])["compound"]
```

```python
df.head()
```

```python
# Average Scores
mean_neu = df["neu"].mean()
mean_neg = df["neg"].mean()
mean_pos = df["pos"].mean()
mean_compound = df["compound"].mean()
print(f"Average Negative Score: {mean_neg:.3f}")
print(f"Average Neutral Score: {mean_neu:.3f}")
print(f"Average Positive Score: {mean_pos:.3f}")
print(f"Average Compound Score: {mean_compound:.3f}")
```
