```python
import matplotlib.pyplot as plt
```

```python
# Followers
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
org = ['SCS', 'MSF']
Followers = [3488, 11803]
ax.bar(org, Followers, color=['green', 'blue'])

for i in range(len(Followers)):
    plt.annotate(str(Followers[i]), xy=(org[i],Followers[i]), ha='center', va='bottom')

ax.set_title('Followers')
ax.set_ylabel('Count')
ax.set_xlabel('Organisation')
plt.show()
```

```python
# Followering
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
org = ['SCS', 'MSF']
Followers = [65, 42]
ax.bar(org, Followers, color=['green', 'blue'])

for i in range(len(Followers)):
    plt.annotate(str(Followers[i]), xy=(org[i],Followers[i]), ha='center', va='bottom')

ax.set_title('Followerings')
ax.set_ylabel('Count')
ax.set_xlabel('Organisation')
plt.show()
```

```python
# Verified Followers

x = ['SCS', 'MSF']
y1 = [46, 64]
y2 = [100-46, 100-64]
  
# plot bars in stack manner
ax1 = plt.bar(x, y1, color='r')
ax2 = plt.bar(x, y2, bottom=y1, color='b')
plt.ylabel('Percentage')
plt.xlabel('Organisation')
plt.legend(['Verified', 'Non-verified'])
plt.title("Verified users percentage in Followers")

for r1, r2 in zip(ax1, ax2):
    h1 = r1.get_height()
    h2 = r2.get_height()
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d" % h1 + "%", ha="center", va="center", color="white", fontsize=16, fontweight="bold")
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d" % h2+ "%", ha="center", va="center", color="white", fontsize=16, fontweight="bold")



plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
org = ['SCS', 'MSF']
Followers = [64.62, 52.38]
followers_count = [42,22]
ax.bar(org, Followers, color=['green', 'blue'])

for i in range(len(Followers)):
    plt.annotate(str(Followers[i]) + f"%, {followers_count[i]}", xy=(org[i],Followers[i]), ha='center', va='bottom')

ax.set_title('Friends')
ax.set_ylabel('Percentage')
ax.set_xlabel('Organisation')
plt.show()
```
